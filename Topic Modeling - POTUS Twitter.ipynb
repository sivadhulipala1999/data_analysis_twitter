The first thing we would like to do with the POTUS tweets data (obtained from https://data.world/socialmediadata/obama-white-house-social-media-potus-tweets) is to extract some topics. 

Topics are groups of words extracted from a given corpus of documents. This would give us an idea of the contents of the corpus

Primary reference: https://towardsdatascience.com/end-to-end-topic-modeling-in-python-latent-dirichlet-allocation-lda-35ce4ed6b3e0 <br/>
How does it work: https://medium.com/analytics-vidhya/latent-dirichelt-allocation-1ec8729589d4 <br/>
Hyperparameter Tuning for number of topics: https://www.machinelearningplus.com/nlp/topic-modeling-gensim-python/#17howtofindtheoptimalnumberoftopicsforlda

```python
import pandas as pd 
import re 
import os 
```

```python
df = pd.read_csv("POTUStweets.csv")
df['text'].head()
```

Punctuations and inconsistent casing could cause problems with the topic modelling process. Hence we remove them for more reliable results.

```python
# Remove punctuation
df['text'] = df['text'].map(lambda x: re.sub("[,\.!?']", '', x))
# Convert the titles to lowercase
df['text'] = df['text'].map(lambda x: x.lower())
df['text'].head()
```

Now we convert the text data to a format which can be fed to the LDA model <br/>
For this we will be tokenizing the sentences and removing the stopwords 

```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['tco', 'http', 'https', 'rt'])

# tokenize 
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
        
# stopword removal         
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]

data = df.text.values.tolist()
data_words = list(sent_to_words(data))
# remove stop words
data_words = remove_stopwords(data_words)
print(data_words[:1][0][:30])
```

The variable data_words is a list consisting of tokens of the input sentences <br/>
For each sentence, a set of tokens would be seen thus making data_words a list of lists 

The tokenized object is now converted to a corpus and a dictionary 

```python
import gensim.corpora as corpora

# Create Dictionary
id2word = corpora.Dictionary(data_words)

# Create Corpus
texts = data_words

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1][0][:30])
```

LDA model is now built and trained on the data. The output of this model would be a list of topics, each being a weighted combination of words.  

```python
from pprint import pprint
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
import pyLDAvis.gensim
import pickle 
import pyLDAvis

# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = os.path.join('./results/ldavis_prepared_'+str(num_topics))

# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself
if 1 == 1:
    LDAvis_prepared = pyLDAvis.gensim.prepare(lda_model, corpus, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)

# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, './results/ldavis_prepared_'+ str(num_topics) +'.html')

LDAvis_prepared
```

# 2nd Attempt: Better Interpretability of Topics

We are now going to do further preprocessing on our input data to have a better clustering of topics. We shall also find the right number of topics as previously we just chose the number 10 randomly. 

Initial preprocessing will be the same in this case too. Read the data, remove punctuations and bring the letter casing to a consistency 

```python
df = pd.read_csv("POTUStweets.csv")

# Remove punctuation
df['text'] = df['text'].map(lambda x: re.sub("[,\.!?']", '', x))
# Convert the titles to lowercase
df['text'] = df['text'].map(lambda x: x.lower())
df['text'].head()
```

We now tokenize the sentences after removing punctuations  

```python
import gensim
from gensim.utils import simple_preprocess

def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data = df.text.values.tolist()
data_words = list(sent_to_words(data))
```

Now, we do something different. Instead of directly feeding the tokens to the LDA model, we form bigrams and trigrams of the tokens. This improves the interpretability of the topics. It also makes the LDA work better in terms of creating the topics. 

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
# NLTK Stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])

# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """Takes in a list of token groups and removes those tokens which are not in the list of allowed POS tags"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
import spacy
import en_core_web_sm

# Remove Stop Words
data_words_nostops = remove_stopwords(data_words)

# Form Bigrams
data_words_bigrams = make_bigrams(data_words_nostops)

# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
nlp = en_core_web_sm.load()

# Do lemmatization keeping only noun, adj, vb, adv
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

print(data_lemmatized[:1][0][:30])
```

```python
import gensim.corpora as corpora

# Create Dictionary - dictionary containing the term against its integer id 
id2word = corpora.Dictionary(data_lemmatized)

# Create Corpus 
texts = data_lemmatized

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1][0][:30]) # shows (token_id, token_count)
```

```python
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=10, 
                                       random_state=100,
                                       chunksize=100,
                                       passes=10,
                                       per_word_topics=True)
```

```python
from pprint import pprint

# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

Now, we calculate the Coherence Score to see how good the model actually was for the given parameter values. 

```python
from gensim.models import CoherenceModel

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score: ', coherence_lda)
```

Since we want to choose a reasonable value for number of topics, we shall calculate the coherence score for different values of this parameter and compare them

```python
# supporting function
def compute_coherence_values(corpus, dictionary, k):
    
    lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=k, 
                                           random_state=100,
                                           chunksize=100,
                                           passes=10,
                                           per_word_topics=True)
    
    coherence_model_lda = CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
    
    return coherence_model_lda.get_coherence()

coherence_vals_dict = {}
for num_topics in range(1, 20): 
    coherence_vals_dict[num_topics] = compute_coherence_values(corpus=corpus, dictionary=id2word, k=num_topics)
```

```python
import matplotlib.pyplot as plt 
import numpy as np

plt.scatter(list(coherence_vals_dict.keys()), list(coherence_vals_dict.values()))
plt.xticks(np.arange(0, 21, step=1))
plt.show()
```

The coherence score seems to be the highest at 17 topics. We shall consider this as our parameter value and build the model 

```python
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=17, 
                                           random_state=100,
                                           chunksize=100,
                                           passes=10,
                                           per_word_topics=True)

from pprint import pprint

# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
